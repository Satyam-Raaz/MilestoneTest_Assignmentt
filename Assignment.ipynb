```python
#Question_7
"""Write a code for  Hotel Management System using OOPS 
Create a Room class that has attributes such as room number, room type, rate, and availability (private)
Implement methods to book a room, check in a guest, and check out a guest
Use encapsulation to hide the room's unique identification number
Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own 
specific attributes and methods."""


class Room:
    def __init__(self, room_number, room_type, rate):
        self.__room_number = room_number  # Private attribute
        self.room_type = room_type
        self.rate = rate
        self.__availability = True  # Room is available by default

    def book_room(self):
        if self.__availability:
            self.__availability = False
            return True
        return False

    def check_in(self):
        if not self.__availability:
            return "Room already occupied."
        else:
            self.__availability = False
            return f"Checked in to room {self.__room_number}."

    def check_out(self):
        if self.__availability:
            return "Room is already available."
        else:
            self.__availability = True
            return f"Checked out from room {self.__room_number}."

    def get_room_info(self):
        return {
            'Room Number': self.__room_number,
            'Room Type': self.room_type,
            'Rate': self.rate,
            'Available': self.__availability
        }
    
class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_jacuzzi=False):
        super().__init__(room_number, 'Suite', rate)
        self.has_jacuzzi = has_jacuzzi

    def get_room_info(self):
        info = super().get_room_info()
        info['Has Jacuzzi'] = self.has_jacuzzi
        return info    
    

class StandardRoom(Room):
    def __init__(self, room_number, rate, is_smoking=False):
        super().__init__(room_number, 'Standard', rate)
        self.is_smoking = is_smoking

    def get_room_info(self):
        info = super().get_room_info()
        info['Is Smoking'] = self.is_smoking
        return info


suite=SuiteRoom(room_number=101,rate=200,has_jacuzzi=True)
Standard=StandardRoom(room_number=102,rate=100,is_smoking=False)

Standard.check_in()
suite.get_room_info()
```

```python
#Question_9
"""Create an Event class that has attributes such as name, date, time, location, and list of attendees (private)
Implement methods to create a new event, add or remove attendees, and get the total number of 
attendees
Use encapsulation to hide the event's unique identification number
Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own 
specific attributes and methods."""

import itertools

class Event:
    _id_counter = itertools.count(1)  # Unique ID counter

    def __init__(self, name, date, time, location):
        self._id = next(Event._id_counter)  # Generate a unique ID
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []  # Private attribute for list of attendees

    def _get_id(self):
        return self._id

    def add_attendee(self, attendee_name):
        if attendee_name not in self.__attendees:
            self.__attendees.append(attendee_name)
            return f"{attendee_name} added to the event."
        else:
            return f"{attendee_name} is already an attendee."

    def remove_attendee(self, attendee_name):
        if attendee_name in self.__attendees:
            self.__attendees.remove(attendee_name)
            return f"{attendee_name} removed from the event."
        else:
            return f"{attendee_name} was not found in the attendee list."

    def get_total_attendees(self):
        return len(self.__attendees)

    def get_event_details(self):
        return {
            "ID": self._get_id(),
            "Name": self.name,
            "Date": self.date,
            "Time": self.time,
            "Location": self.location,
            "Total Attendees": self.get_total_attendees()
        }

class PrivateEvent(Event):
    def __init__(self, name, date, time, location, host):
        super().__init__(name, date, time, location)
        self.host = host

    def get_event_details(self):
        details = super().get_event_details()
        details["Host"] = self.host
        return details

class PublicEvent(Event):
    def __init__(self, name, date, time, location, is_free):
        super().__init__(name, date, time, location)
        self.is_free = is_free

    def get_event_details(self):
        details = super().get_event_details()
        details["Is Free"] = "Yes" if self.is_free else "No"
        return details
    
event1=PrivateEvent("Alice Birthday","2024-09-15","18:00","Alice house","Alice")
event2=PublicEvent("Community fair","2024-10-05","10:00","central park",True)
event1.add_attendee("Bob")
print(event1.get_event_details())
print(event2.get_event_details())

event2.add_attendee("saurav")
event2.add_attendee("ayush")
event2.add_attendee("satyam")
print(event2.get_total_attendees())
event2.remove_attendee("satyam")
print(event2.get_total_attendees())
print(event2.get_event_details())
```

```python
#Question_15> . Write a Python module named file_operations.py with functions for reading, writing, and appending data to a file.
#ans> 
def write_to_file(file_name, data):
    """Write the given data to a file, overwriting any existing content."""
    with open(file_name, 'w') as file:
        file.write(data)

def append_to_file(file_name, data):
    """Append the given data to a file."""
    with open(file_name, 'a') as file:
        file.write(data)

def read_from_file(file_name):
    """Read and return the content of a file."""
    with open(file_name, 'r') as file:
        return file.read()
    
#Example of Using file_operations.py:
import file_operations

# Example usage

# Writing data to a file
file_operations.write_to_file('example.txt', 'Hello, World!\n')

# Appending data to the file
file_operations.append_to_file('example.txt', 'This is an appended line.\n')

# Reading data from the file
content = file_operations.read_from_file('example.txt')

print("File Content:")
print(content)
```

```python
#Question_19
"""Create a Python program that reads a text file named "paragraph.txt" and counts the occurrences of 
each word in the paragraph, displaying the results in alphabetical order."""

import string
def count_word_occurence(filename):
    word_counter=Counter()
    with open(filename,"") as file:
        text=file.read()
        text=text.lower()
        text=text.translate(str.maketrans("","",string.punctuation))
        words=text.split()
        word_counter.update(words)

    return dict(sorted(word_counter.items()))


word_counts=count_word_occurence("paragraph.txt")
if word_counts:
    for word,count in word_counts.items():
        print(f"{word}:{count}")
```

```python
#Question  35> A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the hypothesis that the population standard derivation is 10.5.Hint(Use chi-square distribution).
import scipy.stats as stats

# Given data
n = 25  # Sample size
s = 9.0  # Sample standard deviation
sigma_0 = 10.5  # Hypothesized population standard deviation
alpha = 0.05  # Significance level

# Step 1: Calculate the chi-square test statistic
chi_square_stat = (n - 1) * (s ** 2) / (sigma_0 ** 2)

# Step 2: Determine the critical chi-square values
df = n - 1  # Degrees of freedom
chi_square_critical_right = stats.chi2.ppf(1 - alpha / 2, df)
chi_square_critical_left = stats.chi2.ppf(alpha / 2, df)

# Step 3: Decision making
if  chi_square_stat > chi_square_critical_right:
    print("Reject the null hypothesis")
else:
    print( "Do not reject the null hypothesis")

```

```python
#Quwstion 37> 100 students of a PW IOI obtained the following grades in Data Science paper :
"""Grade :[A, B, C, D, E]
Total Frequency :[15, 17, 30, 22, 16, 100]
 Using the  χ 2 test , examine the hypothesis that the distribution of grades is uniform."""
import numpy as np
import scipy.stats as stats

# Given data
grades = ['A', 'B', 'C', 'D', 'E']
observed_frequencies = np.array([15, 17, 30, 22, 16])
total_students = 100
num_grades = len(grades)

# Calculate expected frequencies for a uniform distribution
expected_frequency = total_students / num_grades
expected_frequencies = np.array([expected_frequency] * num_grades)

# Calculate chi-square test statistic
chi_square_stat, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Determine the critical value and compare
alpha = 0.05  # Significance level
df = num_grades - 1  # Degrees of freedom
chi_square_critical = stats.chi2.ppf(1 - alpha, df)

# Decision
decision = "Reject the null hypothesis" if chi_square_stat > chi_square_critical else "Do not reject the null hypothesis"

# Output the results
print(f"Chi-square statistic: {chi_square_stat:.4f}")
print(f"Critical value: {chi_square_critical:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Decision: {decision}")
```
